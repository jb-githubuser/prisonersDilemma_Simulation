```python
mergedCS_df['Average'] = mergedCS_df.mean(axis=1)
avgCD = mergedCS_df['Average'].sort_values(ascending=False)

plt.style.use('default')

categories = {
    'Game Theoretic': ['TitForTat', 'revTitforTat', 'GrudgePlayer'],
    'Learners (Cumulative Reward)': ['PatternLearningPlayer', 'JointPQLearner', 'Bayesian'],
    'Learners (Individual Reward)': ['QLearningPlayer', 'UCBPlayer'],
    'Patterns': ['PatternPlayer', 'Gullible', 'Sly Fox', 'Evan'],
    'Random': ['Randy']
}

colors = {
    'Game Theoretic': '#DA2C38',
    'Learners (Cumulative Reward)': '#226F54',
    'Learners (Individual Reward)': '#87C38F',
    'Patterns': '#52489C',
    'Random': '#011936'
}

# Create a color mapping for each player
color_map = {}
for category, players in categories.items():
    for player in players:
        color_map[player] = colors[category]

# Extract colors for sorted players
sorted_colors = [color_map[player] for player in avgCD.index]

# Plot the bar plot
plt.figure(figsize=(10, 6))
avgCD.plot(kind='bar', color=sorted_colors)
plt.title('Average Scores Against All Players')
plt.xlabel('Player')
plt.ylabel('Average Score')
plt.xticks(rotation=45, ha='right')

# Create custom legend
handles = [plt.Line2D([0], [0], color=color, lw=4) for color in colors.values()]
labels = colors.keys()
plt.legend(handles, labels, prop={'size': 7})

plt.tight_layout()
plt.show()
```
